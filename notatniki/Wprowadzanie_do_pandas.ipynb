# Wprowadzenie do biblioteki Pandas

[**pandas**](https://pandas.pydata.org/) to popularna biblioteka w języku **Python** służąca do analizy i manipulacji danymi. Zapewnia wygodne struktury danych, takie jak **DataFrame** (tabela podobna do arkusza kalkulacyjnego) oraz **Series** (jednowymiarowa tablica z indeksami).

### Kluczowe funkcje **pandas**:
- **Wczytywanie i zapisywanie danych** (CSV, Excel, SQL, JSON i inne)
- **Filtrowanie, sortowanie i grupowanie danych**
- **Obsługa brakujących wartości**
- **Agregacja i przekształcanie danych**
- **Łączenie różnych zbiorów danych**

# **pandas**  i analiza Big Data

Biblioteka **pandas** odgrywa kluczową rolę w analizie **Big Data**, w fazach **przetwarzania, eksploracji i czyszczenia danych**. Ważnym faktem jest, to że działa w pamięci RAM - podobnie jak np. R i wiele innych tego typu rozwiązań. Jest to kluczowe ograniczenie w analizie naprawdę dużych zbiorów. Ale, w łatwy sposób możną ją skalować poprzez integrację z narzędziami takimi jak **Dask**, **Apache Spark** czy **Modin**. Szczególnie Dask zapewnia łatwą paraleryzację bez konieczności zmiany kodu. Istnieje odmiana skupiona na danych przestrzennych - **Geopandas**.

Główne zastosowania **pandas** to:


### 1. **Wczytywanie i przetwarzanie dużych zbiorów danych**
- Odczyt z plików **CSV, JSON, Excel, SQL, Parquet**
- Obsługa dużych plików poprzez **chunking** (`pd.read_csv(chunksize=n)`)
- Integracja z bazami danych **SQL** (np. PostgreSQL, MySQL)

### 2. **Czyszczenie i transformacja danych**
- Obsługa **brakujących wartości** (`df.fillna()`, `df.dropna()`)
- Usuwanie duplikatów (`df.drop_duplicates()`)
- Konwersja typów danych dla optymalizacji pamięci (`astype()`, `pd.to_datetime()`)

### 3. **Agregacja i analiza statystyczna**
- Grupowanie (`groupby()`, `pivot_table()`)
- Analiza statystyczna (`describe()`, `corr()`, `mean()`, `sum()`)
- Tworzenie nowych kolumn i przekształcenia (`apply()`, `map()`, `lambda`)

### 4. **Łączenie i operacje na dużych zbiorach danych**
- Scalanie zbiorów (`merge()`, `concat()`)
- Operacje na indeksach (`set_index()`, `reset_index()`)

### 5. **Wizualizacja i eksploracja danych**
- Tworzenie wykresów przy użyciu **pandas + Matplotlib** (`df.plot()`)
- Szybka eksploracja danych (`df.head()`, `df.info()`)
